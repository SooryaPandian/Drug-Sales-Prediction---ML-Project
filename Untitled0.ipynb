```python
import pandas as pd
```

```python
path = "/content/drive/MyDrive/ml_intern/drug_sales_prediction_dataset.csv"
df = pd.read_csv(path)
```

```python
from sklearn.model_selection import train_test_split

# Define your feature set (X) and target variable (y)
X = df.drop('SalesVolume', axis=1)  # Features (excluding the target 'SalesVolume')
y = df['SalesVolume']  # Target

# Perform the split (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)

# Check the shape to verify the split
print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
```

```python
# One-Hot Encoding categorical features (e.g., 'Season')
X_train = pd.get_dummies(X_train, columns=['Season'], drop_first=True)
X_test = pd.get_dummies(X_test, columns=['Season'], drop_first=True)

# Align columns to ensure the same structure in train and test sets
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

```python
# Apply One-Hot Encoding to 'Month' (and any other categorical variables, if needed)
X_train = pd.get_dummies(X_train, columns=['Month'], drop_first=True)
X_test = pd.get_dummies(X_test, columns=['Month'], drop_first=True)

# Align train and test sets to ensure they have the same columns
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)

# Now, scale the remaining numeric features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit the scaler on the training data and transform both train and test data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize the model
lr_model = LinearRegression()

# Train the model on the training data
lr_model.fit(X_train_scaled, y_train)

# Make predictions on the test data
y_pred = lr_model.predict(X_test_scaled)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Linear Regression Performance:")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize the Linear Regression model
lr_model = LinearRegression()

# Train the model on the training data
lr_model.fit(X_train_scaled, y_train)

# Make predictions on the test data
y_pred = lr_model.predict(X_test_scaled)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Linear Regression Performance:")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train_scaled, y_train)

# Make predictions on the test data
y_pred_rf = rf_model.predict(X_test_scaled)

# Evaluate the Random Forest model
mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Print the evaluation metrics
print("Random Forest Performance:")
print(f"Mean Absolute Error (MAE): {mae_rf}")
print(f"Mean Squared Error (MSE): {mse_rf}")
print(f"R-squared (R2): {r2_rf}")
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Initialize the Gradient Boosting Regressor
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)

# Train the model
gb_model.fit(X_train_scaled, y_train)

# Make predictions on the test data
y_pred_gb = gb_model.predict(X_test_scaled)

# Evaluate the Gradient Boosting model
mae_gb = mean_absolute_error(y_test, y_pred_gb)
mse_gb = mean_squared_error(y_test, y_pred_gb)
r2_gb = r2_score(y_test, y_pred_gb)

# Print the evaluation metrics
print("Gradient Boosting Performance:")
print(f"Mean Absolute Error (MAE): {mae_gb}")
print(f"Mean Squared Error (MSE): {mse_gb}")
print(f"R-squared (R2): {r2_gb}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Concatenate X_train and y_train for correlation analysis
corr_df = pd.concat([X_train, y_train], axis=1)

# Calculate the correlation matrix
corr_matrix = corr_df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw the heatmap
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Create Price Elasticity feature
X_train['PriceElasticity'] = X_train['CompetitorPrice'] / X_train['AveragePrice']
X_test['PriceElasticity'] = X_test['CompetitorPrice'] / X_test['AveragePrice']

# Create Interaction Features
X_train['MarketingPrescription'] = X_train['MarketingSpend'] * X_train['PrescriptionRate']
X_test['MarketingPrescription'] = X_test['MarketingSpend'] * X_test['PrescriptionRate']
```

```python
# Scale the updated features
X_train_updated = pd.get_dummies(X_train, drop_first=True)
X_test_updated = pd.get_dummies(X_test, drop_first=True)

# Align the train and test sets
X_train_updated, X_test_updated = X_train_updated.align(X_test_updated, join='left', axis=1, fill_value=0)

# Scale the data
X_train_scaled = scaler.fit_transform(X_train_updated)
X_test_scaled = scaler.transform(X_test_updated)

# Train the Gradient Boosting model again
gb_model.fit(X_train_scaled, y_train)

# Make predictions and evaluate
y_pred_gb_updated = gb_model.predict(X_test_scaled)

mae_gb_updated = mean_absolute_error(y_test, y_pred_gb_updated)
mse_gb_updated = mean_squared_error(y_test, y_pred_gb_updated)
r2_gb_updated = r2_score(y_test, y_pred_gb_updated)

# Print the updated evaluation metrics
print("Updated Gradient Boosting Performance:")
print(f"Mean Absolute Error (MAE): {mae_gb_updated}")
print(f"Mean Squared Error (MSE): {mse_gb_updated}")
print(f"R-squared (R2): {r2_gb_updated}")
```
